```python
import numpy

Y = higgs.Label == 's'

X = higgs
del(X["Weight"])
del(X["Label"])

#порежем данные
X_train = X[:150000]
X_test = X[150000:]
Y_train = Y[:150000]
Y_test = Y[150000:]

print X[:5]
```

```python
X.shape
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression().fit(X_train,Y_train)
```

```python
from sklearn.metrics import accuracy_score
Y_pred_train = model.predict(X_train)
Y_pred_test = model.predict(X_test)
print "Обучение:",accuracy_score(Y_train,Y_pred_train)
print "Тест:",accuracy_score(Y_test,Y_pred_test)
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier().fit(X_train,Y_train)
```

```python
Y_pred_train = model.predict(X_train)
Y_pred_test = model.predict(X_test)
print "Обучение:",accuracy_score(Y_train,Y_pred_train)
print "Тест:",accuracy_score(Y_test,Y_pred_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=20).fit(X_train,Y_train)
```

```python
Y_pred_train = model.predict(X_train)
Y_pred_test = model.predict(X_test)
print "Обучение:",accuracy_score(Y_train,Y_pred_train)
print "Тест:",accuracy_score(Y_test,Y_pred_test)
```

```python
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier,AdaBoostClassifier
from rep.metaml import ClassifiersFactory
from rep.estimators import XGBoostClassifier,SklearnClassifier

factory = ClassifiersFactory()


trees = 20

factory = ClassifiersFactory()
factory['rfc'] = SklearnClassifier(RandomForestClassifier(n_estimators= trees))
factory['sgb'] = SklearnClassifier(GradientBoostingClassifier(n_estimators = trees))
factory['ada'] = SklearnClassifier(AdaBoostClassifier(n_estimators=trees))
factory['xgb'] = XGBoostClassifier(n_estimators = trees)
factory.fit(X_train,Y_train)
```

```python
from rep.report.metrics import RocAuc
report = factory.test_on(X_test,Y_test)
learning_curve = report.learning_curve(RocAuc(), steps=1)
learning_curve.plot(new_plot=True)
```
