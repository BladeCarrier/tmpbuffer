#обучение модели

```python
def predict(W,X):
    """получить прогноз h(X|w)"""
    
    size = len(X) #размерность ответа
    
    pred = []
    for i in range(size):
        
        X_i = X.irow(i) #i-тая строчка выборки примеров
    
        pred_i = sum(X_i*W) #прогноз модели для i-той строчки
        
        pred.append(pred_i) #предсказание модели для него
        
    return numpy.array(pred)
```

```python
W =  numpy.array([-0.5,0.5,-0.5,0.5,-0.5,-0.5,0.5,0.5,0.5,-0.5,150])
pred = predict(W,X_train)
print pred.shape
print pred[:10]
print X[:10]
```

```python
def loss(X,Y,W):
    "напиши функцию потерь, J(X,Y,W)"
    
    pred = predict(W, X) #предсказание по модели
    
    error = pred - Y#вектор разностей предсказаний и реальных значений
    
    squared_error = error*error #вектор из квадратов ошибок
    
    mean_squared_error = numpy.average(squared_error) #среднее по ним
    return mean_squared_error
```

```python
mse = loss(X,Y,W)
print mse
```

```python
def gradient(X,Y,W):
    "посчитай вектор градиента функции потерь"
    len_gradient = len (W) #размер вектора градиента
    
    #твои вспомогательные вычисления
    pred = predict(W,X)
    
    clause = (pred  - Y)

    w_sum = numpy.zeros(len(W))
    
    for i in range(len(Y)):
        w_sum += X.irow(i)*clause[i]
    j = 1./len(X) *w_sum
    
    gradient = j
    
    return gradient
```

```python
grad = gradient(X_train,Y_train,W)
print grad.shape
print grad
```

```python
def train_model(X,Y, step = 0.1,n_steps = 100):
    #создадим нулевой вектор весов
    #сколько должно быть весов
    len_w = X.shape[1]
    W = numpy.zeros(len_w)
    
    for i in range(n_steps):
        
        grad =  step*gradient(X,Y,W) #вектор антиградиента
        
        W -= grad #новый вектор весов
        
        # loss = loss(X,Y,W)
        #print loss
    return W
```

#оценка качества модели

```python
#обучи модель на тренировочной выборке
W = train_model(X_train,Y_train, step = 0.1,n_steps = 100)
```
