```python
from sklearn.datasets import load_digits
digits = load_digits(n_class = 10)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
for image in digits.images[:10]:
    plt.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')
    plt.show()
```

```python
images = digits.data
Y = digits.target
print images.shape
```

```python
def extract_features(image):
    features = []
    
    image = image.reshape([8,8])
    
    features.append(numpy.sum(image))
    features.append(numpy.sum(image[3:4,2:6])) #hole in 0
    features.append(numpy.sum(image[2:5,2:5])) #filled center
    features.append(numpy.sum(image[3:4,2:3])) #upper hole in 8
    features.append(numpy.sum(image[2:5,5:6])) #lower hole in 8
    features.append(numpy.sum(image[:4,:])/(numpy.sum(image[4:,:])+0.01)) #brightness left/right
    features.append(numpy.sum(image[:,:4])/(numpy.sum(image[:,4:])+0.01)) #brightness upper/lower
    features.append(numpy.sum(image[:4,0:2])/(numpy.sum(image[4:,0:2])+0.01)) #brightness left/right, up
    features.append(numpy.sum(image[:4,3:4])/(numpy.sum(image[4:,3:4])+0.01)) #brightness left/right, middle
    features.append(numpy.sum(image[:4,5:8])/(numpy.sum(image[4:,5:8])+0.01)) #brightness left/right, bottom
    
    return features
```

```python
import numpy
X = []
for i in range(images.shape[0]):
    X.append(extract_features(images[i]))
X = numpy.array(X)
```

```python
print X.shape
```

```python
from sklearn.cross_validation import train_test_split
Xtrain,Xtest,Ytrain,Ytest = train_test_split(X,Y,train_size = 0.5)
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(Xtrain,Ytrain)
```

```python
Ypred = model.predict(Xtest)
```

```python
from sklearn.metrics import accuracy_score
print "test:",accuracy_score(Ypred,Ytest)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(Xtrain,Ytrain)
```

```python
Ypred = model.predict(Xtest)
print "test:",accuracy_score(Ypred,Ytest)
```
